```python
import facebook
import requests
import api_key
import json
# with open('data.json', 'w') as fp:
#     json.dump(data, fp)

topleft = 50.104718, 14.373066
bottomright = 50.068257, 14.447732

import numpy as np

n = 12 # n^2 requests
xx = np.linspace(topleft[0], bottomright[0], n)
yy = np.linspace(topleft[1], bottomright[1], n)
grid = [(x,y) for x in xx for y in yy]
```

```python
ACCESS_TOKEN = api_key.FB_KEY
graph = facebook.GraphAPI(ACCESS_TOKEN)
# friends = graph.get_connections("me","friends")

furl = "https://graph.facebook.com/v2.8/"
# search?q=museum&type=place&center=50.0866089,14.4207597&distance=1000&fields=id,name,location,checkins
req = "search?type=place&center={},{}&distance=350&fields=id,name,checkins,location,fan_count"
tok = "&access_token="+ACCESS_TOKEN

def url(gps):
    lat,lon=gps
    return furl + req.format(lat, lon) + tok
```
